```python
# Write some code that provides a graph laplacian for an Erdos-Renyi graph
import networkx as nx
import numpy as np
import scipy.sparse as sp

def erdos_renyi_graph_laplacian(n, p):
    G = nx.erdos_renyi_graph(n, p)
    A = nx.adjacency_matrix(G)
    D = sp.diags(np.array(A.sum(axis=1)).flatten())
    L = D - A
    return L

# test it
L = erdos_renyi_graph_laplacian(10, 0.5)
print(L)

# display it as a graph
import matplotlib.pyplot as plt
G = nx.erdos_renyi_graph(10, 0.5)
nx.draw(G, with_labels=True)
plt.show()

# perform a random walk on the graph and visually depict it
import random
def random_walk(G, start, length):
    walk = [start]
    for i in range(length):
        neighbors = list(G.neighbors(walk[-1]))
        walk.append(random.choice(neighbors))
    return walk

walk = random_walk(G, 0, 10)
print(walk)
nx.draw(G, with_labels=True)
nx.draw_networkx_nodes(G, nodelist=walk, node_color="red")
plt.show()
```
